```python
import pandas as pd
```

```python
df = pd.read_csv('Hyderabad_scraped_car_data.csv')
df
```

```python
# Clean the 'Price' column to convert it into integer format
def clean_price(price):
    # Remove '₹', 'L', and commas, then convert to float and scale lakhs to integer
    price = price.replace('₹', '').replace('L', '').replace(',', '').strip()
    return int(float(price) * 100000)  # Scale lakhs to integer

# Apply the cleaning function to the 'Price' column
df['Price'] = df['Price'].apply(clean_price)
```

```python
df
```

```python
df.to_csv('car_data.csv', index=False)
```
